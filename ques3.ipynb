```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
myimage = cv2.imread("young messi.jpeg",0)  #reading image in grayscale
```

```python
#NEGATIVE INTENSITY LEVEL TRANSFORMATION

h,w = myimage.shape
max = -1
for i in range(h):
    for j in range(w):
        if myimage[i][j]>max:
            max = myimage[i][j]
negative = max-myimage
```

```python
plt.imshow(myimage, cmap='gray')
plt.title('Original Image'), plt.axis('off')
```

```python
plt.imshow(negative, cmap='gray')
plt.title('Negative Image'), plt.axis('off')
```
